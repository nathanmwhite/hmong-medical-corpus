```python
errors = {}
for item in filenames:
    f = open(os.path.join(anatomy_folder, item), 'r')
    data_f = [[word.split('▹')[0] for word in line.strip().split(' ')] for line in f.readlines()]
    g = open(os.path.join(disease_folder, item), 'r')
    data_g = [[word.split('▹')[0] for word in line.strip().split(' ')] for line in g.readlines()]
    h = open(os.path.join(symptom_folder, item), 'r')
    data_h = [[word.split('▹')[0] for word in line.strip().split(' ')] for line in h.readlines()]
    for i, line in enumerate(data_f):
        try:
            if not (data_f[i] == data_g[i] and data_g[i] == data_h[i]):
                for j, element in enumerate(data_f[i]):
                    if data_f[i][j] == data_g[i][j] and data_g[i][j] != data_h[i][j]:
                        errors[item] = [i] + [j] + ['symptom_disagrees'] + [data_h[i][j]] + data_h[i]
                    elif data_f[i][j] == data_h[i][j] and data_f[i][j] != data_g[i][j]:
                        errors[item] = [i] + [j] + ['disease_disagrees'] + [data_g[i][j]] + data_g[i]
                    elif data_g[i][j] == data_h[i][j] and data_g[i][j] != data_f[i][j]:
                        errors[item] = [i] + [j] + ['anatomy_disagrees'] + [data_f[i][j]] + data_f[i]
                    elif data_f[i][j] != data_g[i][j] and data_f[i][j] != data_h[i][j]:
                        errors[item] = [i] + [j] + ['all_disagree'] + [data_f[i][j]] + data_f[i]
        except IndexError:
            errors[item] = ['index error'] + [len(data_f[i])] + [len(data_g[i])] + [len(data_h[i])] +  data_f[i]
    f.close()
    g.close()
    h.close()
```

```python
from itertools import zip_longest
results = {}
for item in filenames:
    f = open(os.path.join(anatomy_folder, item), 'r')
    data_f = [[word.split('▹')[0] for word in line.strip().split(' ')] for line in f.readlines()]
    g = open(os.path.join(disease_folder, item), 'r')
    data_g = [[word.split('▹')[0] for word in line.strip().split(' ')] for line in g.readlines()]
    h = open(os.path.join(symptom_folder, item), 'r')
    data_h = [[word.split('▹')[0] for word in line.strip().split(' ')] for line in h.readlines()]
    data = []
    for i, line in enumerate(data_f):
        data.append(list(zip_longest(line, data_g[i], data_h[i], fillvalue='???')))
    results[item] = data
        
```

```python
finds = []
for item in results.keys():
    for i, line in enumerate(results[item]):
        for j, word in enumerate(line):
            try:
                if not (word[0].lower() == word[1].lower() and word[1].lower() == word[2].lower()):
                    finds.append([item, i, j, word])
                    break
            except IndexError:
                finds.append([item, i, j, word])
```

```python
results = {}
errors = {}
for item in filenames:
    f = open(os.path.join(anatomy_folder, item), 'r')
    content_f = [[word.split('▹') for word in line.strip().split(' ')] for line in f.readlines()]
    data_f = [[word[0] for word in line] for line in content_f]
    tags_f = [[word[1] for word in line] for line in content_f]
    f.close()
    g = open(os.path.join(disease_folder, item), 'r')
    content_g = [[word.split('▹') for word in line.strip().split(' ')] for line in g.readlines()]
    data_g = [[word[0] for word in line] for line in content_g]
    tags_g = [[word[1] for word in line] for line in content_g]
    g.close()
    h = open(os.path.join(symptom_folder, item), 'r')
    content_h = [[word.split('▹') for word in line.strip().split(' ')] for line in h.readlines()]
    data_h = [[word[0] for word in line] for line in content_h]
    tags_h = [[word[1] for word in line] for line in content_h]
    h.close()
    data = []
    for i, line in enumerate(data_f):
        out = []
        current = list(zip(line, data_g[i], data_h[i]))
        for j, word in enumerate(current):
            if word[0] == word[1] or word[0] == word[2]:
                out.append(word[0])
            elif word[1] == word[2]:
                out.append(word[1])
            else:
                to_add = (i, j, word)
                if item in errors.keys():
                    errors[item].append(to_add)
                else:
                    errors[item] = [to_add]
        data.append(out)
    results[item] = data

    f = open(os.path.join(anatomy_folder, item), 'w')
    g = open(os.path.join(disease_folder, item), 'w')
    h = open(os.path.join(symptom_folder, item), 'w')
    for k, line in enumerate(results[item]):
        if k < len(results[item]) - 1:
            f.write(' '.join(['▹'.join(word) for word in zip(line, tags_f[k])]) + '\n')
            g.write(' '.join(['▹'.join(word) for word in zip(line, tags_g[k])]) + '\n')
            h.write(' '.join(['▹'.join(word) for word in zip(line, tags_h[k])]) + '\n')
        else:
            f.write(' '.join(['▹'.join(word) for word in zip(line, tags_f[k])]))
            g.write(' '.join(['▹'.join(word) for word in zip(line, tags_g[k])]))
            h.write(' '.join(['▹'.join(word) for word in zip(line, tags_h[k])]))            
    f.close()
    g.close()
    h.close()
```
