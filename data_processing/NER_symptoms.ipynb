```python
filenames = glob.glob('p*h.txt')
```

```python
data = {}
for file in filenames:
    data[file] = PCR('.', file).sents()
```

```python
# removes documents that do not have bullet lists
candidate_files = sorted(k for (k,v) in data.items() if 'Fact' not in list(chain(*v)))
other_files = sorted(k for (k,v) in data.items() if 'Fact' in list(chain(*v)))
```

```python
sents = []
for item in candidate_files:
    for i, sent in enumerate(data[item]):
        try:
            if 'tsos' in sent and '*' in data[item][i + 1]:
                sents.append(data[item][i + 1])
        except IndexError:
            continue
```

```python
print(sents)
```

```python
# remove detritus introduced by PCR process
# Once a sequence of 'word', 'Word' is reached without an asterisk, the symptoms are over
symptoms = []
for sent in sents:
    current = False
    current_symptom = []
    for word in sent:
        if current == True:
            if word == '*':
                symptoms.append(current_symptom)
                current_symptom = []
            elif word[0] in string.ascii_uppercase and len(current_symptom) > 0:
                symptoms.append(current_symptom)
                break
            else:
                current_symptom.append(word)
        else: # current == False
            if word == '*':
                current = True
```

```python
len(symptoms)
```

```python
unique_candidates = list(set(tuple(s) for s in symptoms))
```

```python
print(unique_candidates)
```

```python
f = open('unique_candidates.txt', 'w')
for item in unique_candidates:
    f.write(' '.join(item))
    f.write('\n')
f.close()
```

```python
# process other file docs such that they are TokenSearcher searchable
# filter on the first step such that the sentences must contain one of 'yam', 'muaj', 'xws', 'yuav', or 'yog'
#    for efficiency: other sentences do not need to be converted, only likely candidates
# these next two can be condensed down
unprocessed_hits = []
for file in [file for file in filenames if file not in candidate_files]:
    current_question = ''
    correct_question = False
    for sent in data[file]:
        if '?' in sent:
            current_question = sent
            if 'tsos' in current_question and 'mob' in current_question and 'zoo' in current_question \
            and 'li' in current_question and 'cas' in current_question:
                correct_question = True
            else:
                correct_question = False
        elif correct_question == True:
            if 'yam' in sent or 'muaj' in sent or 'xws' in sent or 'yuav' in sent or 'yog' in sent:
                if (',' in sent or '*' in sent) and 'thiab' in sent:
                    unprocessed_hits.append(sent)
        else:
            continue
```

```python
searchables = ['(?:<muaj>|<yam>|<xws>)<li><no>?<[,:]>(<.*>*)', '<mob><.*>{0,2}<mas><yuav>(?!<tsum>)(<.*>*)<[\.\?]>', '<mob><.*>?<kuj><yuav>(<.*>*)<[\.\?]>', '<mob><no>?<(?:yuav|yog)>(?!<tsum>)(?!<hais>)(?!<kis>)(?!<siv>)(?!<cov>)(<.*>*)<[\.\?]>']
# the third and fourth approaches produce mostly garbage; the second produces the best
# how to eliminate the garbage?
selected_sents_by_searchable = [[], [], [], []]
for sent in unprocessed_hits:
    ts = TS(sent)
    for i, re_string in enumerate(searchables):
        found = ts.findall(re_string)
        if len(found) > 0 and (',' in found[0] or '*' in found[0]):
            selected_sents_by_searchable[i].append(found)
```

```python
f = open('unstructured_finds.txt', 'w')
for i, searchable in enumerate(selected_sents_by_searchable):
    f.write(searchables[i])
    f.write('\n')
    for item in searchable:
        f.write(' '.join(item[0]))
        f.write('\r\n\r\n')
f.close()
```
